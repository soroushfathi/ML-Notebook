```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data = data.drop(['Unnamed: 0', 'id'], axis=1).dropna()
data.head()
```

```python
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])
```

```python
X = data.drop('satisfaction', axis=1)
y = data['satisfaction']
```

```python
from sklearn.linear_model import Perceptron
per = Perceptron(random_state=0)
per.fit(X.values, y.values)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y.values, per.predict(X.values))
```
